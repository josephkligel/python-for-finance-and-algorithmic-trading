```python
import matplotlib.dates as dates
```

```python
mcdon['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=(0,50))
```

```python
idx = mcdon.loc['2007-01':'2007-05-01'].index
```

```python
stock = mcdon.loc['2007-01':'2007-05-01']['Adj. Close']
```

```python
stock
```

```python
fig,ax = plt.subplots()
ax.plot(idx,stock)
ax.yaxis.grid(True)
ax.xaxis.grid(True)
fig.autofmt_xdate()

#Locating
ax.xaxis.set_major_locator(dates.MonthLocator())
#Formatting
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=6))
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))
```
